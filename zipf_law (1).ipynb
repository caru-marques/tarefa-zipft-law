Para esta tarefa, foi escolhido um texto do site indicado pelo professor Rogério https://www.gutenberg.org

Link de acesso no Github - https://github.com/caru-marques/zipf-law



```python
  
import urllib.request  

texto = []

for line in urllib.request.urlopen('https://www.gutenberg.org/cache/epub/66456/pg66456.txt'): 
    texto.append(line.decode('utf-8'))

for i in range(len(texto)):
    texto[i] = texto[i].lower() 
    texto[i] = texto[i].replace('\n','')
    texto[i] = texto[i].replace('.','')
    texto[i] = texto[i].replace(',','')
    texto[i] = texto[i].replace('(','')
    texto[i] = texto[i].replace(')','')
    texto[i] = texto[i].replace('?','')
    texto[i] = texto[i].replace('\'','')
```

```python
mydict = {}                             

for line in texto:

  line = line.lower()                   
  words = line.split()                  


  for word in words:                      
    if word not in mydict.keys():         
      mydict[word] = 1                    
    else:                                 
      mydict[word] =  mydict[word] + 1    

print(mydict)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.DataFrame(mydict.items(),  columns=['word', 'count']).sort_values('count',ascending=False)
df = df[df['count'] > 4] 

plt.figure(figsize=(24,10))
mpl.style.use(['seaborn'])
sns.barplot(x=df.word,y=df['count'])
plt.xticks(rotation=90) 

plt.show()
```

Com esta atividade, foi possível observa que a lei de Zipf foi funcionante com o texto escolhido.
